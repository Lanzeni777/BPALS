```python
def get_historical_data():
    self = kl_db.dp
    d = self.tickers_manager.get_all_stocks_tickers()
    d.index = d["Ticker"]
    st, ref, dfs = [], list(d["Ticker"]), {}
    for file in os.listdir("../DataBase/XLSX"):
        df = pd.read_csv(f"../DataBase/XLSX/{file}")
        tck = file.split()[0]
        df["Ticker"] = tck
        df["Description"] = self.tickers_manager.get_description(tck)
        st.append(tck)
        
        # clean the df
        df.columns = ['Date', 'Close', 'Open', 'High', 'Low', 'Volume', 'Change', 'Ticker', 'Description']
        df = df.drop("Change", axis=1)
        df = df.fillna("0.0")
        df["Volume"] = [int(float(str(df["Volume"][j]).replace("K", "").replace("M", "")))*1000 for j in range(len(list(df["Volume"])))]
        df["Date"] = self.make_date([f"{dd.split('/')[-1]}-{dd.split('/')[0]}-{dd.split('/')[1]}" for dd in df["Date"]])
        df.index = df["Date"]
        dfs[tck] = self.clean_and_convert(df)
        if not tck in ref:
            print("to remove  ", tck)
    for e in ref:
        if not e in st:
            print("to add ", e, d.loc[e, "Description"])
    return dfs
            
```

```python
def merge_dfs(dfs):
    merged = []
    for ticker, df in dfs.items():
        df = df.copy()
        df['Ticker'] = ticker
        df['Description'] = kl_db.dp.tickers_manager.get_description(ticker)
        merged.append(df)

    merged_df = pd.concat(merged)
    merged_df.reset_index(inplace=True)
    merged_df.set_index(['Date', 'Ticker'], inplace=True)
    return merged_df

def insert_3m_histo_data(merged_df):
    # === 1. Connect to SQLite DB ===
    merged_df = merge_dfs(kl_db.dp.get_3m_hist_data())
    kl_db._connect()
    # === 2. Load existing date+ticker pairs to avoid duplicates ===
    existing = pd.read_sql("SELECT Date, Ticker FROM stock_prices", kl_db.conn)
    existing['Date'] = pd.to_datetime(existing['Date'], dayfirst=True, errors='coerce')

    # === 3. Assume merged_df already exists (from prior merge) ===
    # Ensure index is reset and column names are clean
    merged_df = merged_df.reset_index()

    # === 4. Filter out already existing entries ===
    merged_df['Date'] = pd.to_datetime(merged_df['Date'], dayfirst=True, errors='coerce')
    check_df = merged_df.merge(existing, on=['Date', 'Ticker'], how='left', indicator=True)
    #display(check_df.head())
    merged_df = check_df[check_df['_merge'] == 'left_only'].drop(columns=['_merge', 'Volume(FCFA)'])

    # === 5. Insert into 'prices' table ===
    if not merged_df.empty:
        merged_df.to_sql('stock_prices', kl_db.conn, if_exists='append', index=False)
        print(f"Inserted {len(merged_df)} new rows into 'prices' table.")
    else:
        print("No new data to insert. All entries already exist.")

    kl_db.close()


def get_last_prices():
    self = kl_db
    self._connect()
    daily_df1 = kl_db.dp.get_table_from_url("sikafinance", "price")
    daily_df2 = kl_db.dp.get_table_from_url("sikafinance", "variation")
    # Ensure ticker mapping is available
    # tickers_ref = self.dp.tickers_manager.get_all_brvm_tickers() #pd.read_sql("SELECT Ticker, Description FROM tickers", kl_db.conn)
    # tickers_ref['Description'] = tickers_ref['Description'].str.upper()

    daily_df1["Volume"] = daily_df2["Volume"] #.merge(daily_df2, left_on="Description", right_on="Description", how="left")
    daily_df1["Ticker"] = daily_df1.index
    df = daily_df1.fillna("-")
    df["Date"] = self.make_date(df["Date"])
    #display(df)
    # Final formatting
    #df = df[['Ticker', 'Open', 'High', 'Low', 'Close']].copy()
    # df.columns = ['ticker', 'open', 'high', 'low', 'close']
    #df['Date'] = pd.to_datetime(daily_date, dayfirst=True, errors='coerce')

    # Avoid duplicates
    existing = pd.read_sql(f"SELECT Date, Ticker FROM stock_prices WHERE Date = '{list(df['Date'])[0]}'", self.conn)
    existing['Date'] = self.make_date(existing['Date'])
    check = df.merge(existing, on=['Date', 'Ticker'], how='left', indicator=True)
    #display(existing.head(65))
    #display(check.head(65))
    #print(list(check["_merge"]))
    df = check[check['_merge'] == 'left_only'].drop(columns=['_merge', 'Volume(FCFA)'])
    return df
    
    
def insert_daily_prices():
    df = get_last_prices()
    if not df.empty:
        df.to_sql('stock_prices', self.conn, if_exists='append', index=False)
        print(f"Inserted {len(df)} new daily rows into 'prices' table.")
    else:
        print("No new daily data to insert. All entries already exist.")

    self.close()

# Example usage (replace daily_df and daily_date with real data)
# insert_daily_prices(daily_df, '30/05/2025')
def check_value(table, date, ticker):
    self = kl_db
    self._connect()
    self.cursor.execute(f"SELECT EXISTS(SELECT 1 FROM {table} WHERE Date='{date}' AND Ticker='{ticker}')")
    exists = self.cursor.fetchone()[0]
    print(f"({date}, {ticker}) Exists in {table}" if exists else f"({date}, {ticker}) is missing in {table}")
    self.close()
    return exists

def insert_if_not_exists(table, row):
    self = kl_db
    exists = check_value(table, row["Date"], row["Ticker"])
    
    if not exists:
        keys = ', '.join(row.keys())
        placeholders = ', '.join(['?'] * len(row))
        values = tuple(row.values())

        sql = f"INSERT OR IGNORE INTO {table} ({keys}) VALUES ({placeholders})"
        print(sql)
        kl_db._connect()
        kl_db.cursor.execute(sql, values)
        kl_db.conn.commit()
        kl_db.close()
        print("Done")
    else:
        print(f"{row} Already exists")


import pandas as pd

def clean_and_convert(df):
    df_clean = df.copy()
    for col in df_clean.columns:
        if df_clean[col].dtype == object:
            # Supprime espaces insécables, normaux, et convertit en int si possible
            df_clean[col] = (
                df_clean[col]
                .astype(str)
                .str.replace('\xa0', '', regex=False)
                .str.replace(' ', '', regex=False)
                .str.strip()
            )
            # Essaie de convertir en entier si possible
            try:
                df_clean[col] = df_clean[col].astype(int)
            except ValueError:
                pass  # Ignore si conversion impossible
    return df_clean
```

```python
import requests
from bs4 import BeautifulSoup
import os
import pdfplumber

def get_boc_on_nsia():
    # URL de la page contenant les liens vers les bulletins PDF
    base_url = "https://nsiafinance.com/donnees-de-marche/"

    # Effectuer une requête GET pour obtenir le contenu de la page
    response = requests.get(base_url)
    response.raise_for_status()  # Vérifie que la requête a réussi

    # Analyser le contenu HTML de la page
    soup = BeautifulSoup(response.text, "html.parser")

    # Trouver tous les liens vers des fichiers PDF
    pdf_links = []
    for link in soup.find_all("a", href=True):
        href = link["href"]
        if href.lower().endswith(".pdf"):
            pdf_links.append(href)

    # Vérifier si des liens PDF ont été trouvés
    if not pdf_links:
        print("Aucun lien PDF trouvé sur la page.")
    else:
        # Prendre le premier lien PDF trouvé (le plus récent)
        print(pdf_links)
        pdf_url = pdf_links[0]
        # Si le lien est relatif, le convertir en lien absolu
        if not pdf_url.startswith("http"):
            pdf_url = requests.compat.urljoin(base_url, pdf_url)

        # Télécharger le fichier PDF
        pdf_response = requests.get(pdf_url)
        pdf_response.raise_for_status()

        # Extraire le nom du fichier à partir de l'URL
        filename = f"../PDF/{os.path.basename(pdf_url)}"

        # Enregistrer le fichier PDF localement
        with open(filename, "wb") as f:
            f.write(pdf_response.content)

        print(f"Fichier téléchargé : {filename}")
    return filename
```

```python
def pdfl(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        dfs = []
        for page in pdf.pages:
            tables = page.extract_tables()
            for table in tables:
                df = pd.DataFrame(table[1:], columns=table[0])
                dfs.append(df)
    excel_file_path = pdf_path
    excel_file_path = excel_file_path.replace(".pdf", ".xlsx")
    with pd.ExcelWriter(excel_file_path) as writer:
        # Iterate over dictionary items and write each DataFrame to a separate sheet
        for i, df in enumerate(dfs):
            df.to_excel(writer, sheet_name=f"Table {i}", index=True)
```

```python
pdfl(get_boc_on_nsia())
```

```python
def get_data_saved(excel_file_path):
    
    # excel_file_path = f"{path()}/Data/Stocks/{symbol}/{d.strftime('%Y-%m-%d')}.xlsx"
    xls = pd.ExcelFile(excel_file_path)
    # Get the sheet names
    sheet_names = xls.sheet_names
    # Initialize a dictionary to store DataFrames
    data = {}
    # Iterate over sheet names and read each sheet into a DataFrame
    for sheet_name in sheet_names:
        data[sheet_name] = pd.read_excel(xls, sheet_name=sheet_name)
        data[sheet_name].index = data[sheet_name]["Unnamed: 0"]

    return data
```
