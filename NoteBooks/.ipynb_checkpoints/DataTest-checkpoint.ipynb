```python
import requests
from bs4 import BeautifulSoup
import os
import pdfplumber

def get_boc_on_nsia():
    # URL de la page contenant les liens vers les bulletins PDF
    base_url = "https://nsiafinance.com/donnees-de-marche/"

    # Effectuer une requête GET pour obtenir le contenu de la page
    response = requests.get(base_url)
    response.raise_for_status()  # Vérifie que la requête a réussi

    # Analyser le contenu HTML de la page
    soup = BeautifulSoup(response.text, "html.parser")

    # Trouver tous les liens vers des fichiers PDF
    pdf_links = []
    for link in soup.find_all("a", href=True):
        href = link["href"]
        if href.lower().endswith(".pdf"):
            pdf_links.append(href)

    # Vérifier si des liens PDF ont été trouvés
    if not pdf_links:
        print("Aucun lien PDF trouvé sur la page.")
    else:
        # Prendre le premier lien PDF trouvé (le plus récent)
        print(pdf_links)
        pdf_url = pdf_links[0]
        # Si le lien est relatif, le convertir en lien absolu
        if not pdf_url.startswith("http"):
            pdf_url = requests.compat.urljoin(base_url, pdf_url)

        # Télécharger le fichier PDF
        pdf_response = requests.get(pdf_url)
        pdf_response.raise_for_status()

        # Extraire le nom du fichier à partir de l'URL
        filename = f"../PDF/{os.path.basename(pdf_url)}"

        # Enregistrer le fichier PDF localement
        with open(filename, "wb") as f:
            f.write(pdf_response.content)

        print(f"Fichier téléchargé : {filename}")
    return filename
```

```python
def pdfl(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        dfs = []
        for page in pdf.pages:
            tables = page.extract_tables()
            for table in tables:
                df = pd.DataFrame(table[1:], columns=table[0])
                dfs.append(df)
    excel_file_path = pdf_path
    excel_file_path = excel_file_path.replace(".pdf", ".xlsx")
    with pd.ExcelWriter(excel_file_path) as writer:
        # Iterate over dictionary items and write each DataFrame to a separate sheet
        for i, df in enumerate(dfs):
            df.to_excel(writer, sheet_name=f"Table {i}", index=True)
```

```python
def get_data_saved(excel_file_path):
    
    # excel_file_path = f"{path()}/Data/Stocks/{symbol}/{d.strftime('%Y-%m-%d')}.xlsx"
    xls = pd.ExcelFile(excel_file_path)
    # Get the sheet names
    sheet_names = xls.sheet_names
    # Initialize a dictionary to store DataFrames
    data = {}
    # Iterate over sheet names and read each sheet into a DataFrame
    for sheet_name in sheet_names:
        data[sheet_name] = pd.read_excel(xls, sheet_name=sheet_name)
        data[sheet_name].index = data[sheet_name]["Unnamed: 0"]

    return data
```

```python
import os
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import pandas as pd
import sqlite3

class BRVMBulletinManager:
    def __init__(self, storage_dir="bulletins", base_url="https://nsiafinance.com/donnees-de-marche/", db_path="KLBrvm_BataBase.db"):
        self.storage_dir = storage_dir
        self.base_url = base_url
        self.db_path = db_path
        os.makedirs(self.storage_dir, exist_ok=True)
        self._init_db()

    def _init_db(self):
        conn = sqlite3.connect(self.db_path)
        cursor = conn.cursor()
        cursor.execute('''
            CREATE TABLE IF NOT EXISTS bulletin_data (
                Date TEXT,
                Section TEXT,
                Field TEXT,
                Value TEXT,
                PRIMARY KEY (Date, Section, Field)
            )
        ''')
        conn.commit()
        conn.close()

    def fetch_pdf_links(self):
        response = requests.get(self.base_url)
        soup = BeautifulSoup(response.text, 'html.parser')
        links = [a['href'] for a in soup.find_all('a', href=True) if a['href'].lower().endswith('.pdf')]
        full_links = [(self._extract_date_from_filename(link), os.path.basename(link), link) for link in links]
        return full_links

    def _extract_date_from_filename(self, filename):
        try:
            parts = filename.split('_')
            date_part = parts[-1].replace(".pdf", "")[:8]
            return datetime.strptime(date_part, "%Y%m%d").strftime("%Y-%m-%d")
        except:
            return None

    def download_bulletins(self, links):
        saved_files = []
        for date_str, fname, url in links:
            local_path = os.path.join(self.storage_dir, fname)
            if not os.path.exists(local_path):
                try:
                    print(f"Downloading {fname}...")
                    res = requests.get(url)
                    with open(local_path, 'wb') as f:
                        f.write(res.content)
                    saved_files.append((date_str, fname))
                except Exception as e:
                    print(f"Failed to download {url}: {e}")
        return saved_files

    def list_local_bulletins(self):
        return [f for f in os.listdir(self.storage_dir) if f.endswith('.pdf')]

    def get_bulletin_path(self, date):
        for file in self.list_local_bulletins():
            if date.replace('-', '') in file:
                return os.path.join(self.storage_dir, file)
        return None

    def parse_bulletin_text(self, filepath):
        from PyPDF2 import PdfReader
        reader = PdfReader(filepath)
        full_text = "\n".join([page.extract_text() for page in reader.pages if page.extract_text()])
        return full_text

    def structure_info(self, date, raw_text):
        # Placeholder logic – should parse actual structure into dataframe-like rows
        # For demo, extract lines containing ':'
        records = []
        for line in raw_text.splitlines():
            if ':' in line:
                parts = line.split(':', 1)
                records.append((date, "generic", parts[0].strip(), parts[1].strip()))
        return pd.DataFrame(records, columns=["Date", "Section", "Field", "Value"])

    def store_structured_info(self, df):
        conn = sqlite3.connect(self.db_path)
        df.to_sql("bulletin_data", conn, if_exists='append', index=False)
        conn.commit()
        conn.close()

    def process_new_bulletins(self):
        links = self.fetch_pdf_links()
        downloaded = self.download_bulletins(links)
        for date_str, fname in downloaded:
            path = os.path.join(self.storage_dir, fname)
            raw_text = self.parse_bulletin_text(path)
            structured_df = self.structure_info(date_str, raw_text)
            self.store_structured_info(structured_df)

    def load_bulletin_from_db(self, date):
        conn = sqlite3.connect(self.db_path)
        df = pd.read_sql("SELECT * FROM bulletin_data WHERE Date = ?", conn, params=(date,))
        conn.close()
        return df
```

```python
import os
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import pandas as pd
import sqlite3
import re

class BRVMBulletinManager:
    def __init__(self, storage_dir="bulletins", base_url="https://nsiafinance.com/donnees-de-marche/", db_path="KLBrvm_BataBase.db"):
        self.storage_dir = storage_dir
        self.base_url = base_url
        self.db_path = db_path
        os.makedirs(self.storage_dir, exist_ok=True)
        self._init_db()

    def _init_db(self):
        conn = sqlite3.connect(self.db_path)
        cursor = conn.cursor()
        cursor.execute('''
            CREATE TABLE IF NOT EXISTS bulletin_data (
                Date TEXT,
                Section TEXT,
                Field TEXT,
                Value TEXT,
                PRIMARY KEY (Date, Section, Field)
            )
        ''')
        cursor.execute('''
            CREATE TABLE IF NOT EXISTS bulletin_table_data (
                Date TEXT,
                Ticker TEXT,
                Price REAL,
                Variation_J REAL,
                Variation_Y REAL,
                PER REAL,
                Dividend REAL,
                PRIMARY KEY (Date, Ticker)
            )
        ''')
        conn.commit()
        conn.close()

    def fetch_pdf_links(self):
        response = requests.get(self.base_url)
        soup = BeautifulSoup(response.text, 'html.parser')
        links = [a['href'] for a in soup.find_all('a', href=True) if a['href'].lower().endswith('.pdf')]
        full_links = [(self._extract_date_from_filename(link), os.path.basename(link), link) for link in links]
        return full_links

    def _extract_date_from_filename(self, filename):
        try:
            parts = filename.split('_')
            date_part = parts[-1].replace(".pdf", "")[:8]
            return datetime.strptime(date_part, "%Y%m%d").strftime("%Y-%m-%d")
        except:
            return None

    def download_bulletins(self, links):
        saved_files = []
        for date_str, fname, url in links:
            local_path = os.path.join(self.storage_dir, fname)
            if not os.path.exists(local_path):
                try:
                    print(f"Downloading {fname}...")
                    res = requests.get(url)
                    with open(local_path, 'wb') as f:
                        f.write(res.content)
                    saved_files.append((date_str, fname))
                except Exception as e:
                    print(f"Failed to download {url}: {e}")
        return saved_files

    def list_local_bulletins(self):
        return [f for f in os.listdir(self.storage_dir) if f.endswith('.pdf')]

    def get_bulletin_path(self, date):
        for file in self.list_local_bulletins():
            if date.replace('-', '') in file:
                return os.path.join(self.storage_dir, file)
        return None

    def parse_bulletin_text(self, filepath):
        from PyPDF2 import PdfReader
        reader = PdfReader(filepath)
        full_text = "\n".join([page.extract_text() for page in reader.pages if page.extract_text()])
        return full_text

    def structure_info(self, date, raw_text):
        records = []
        section = None

        for line in raw_text.splitlines():
            line = line.strip()
            if not line:
                continue

            if "BRVM COMPOSITE" in line.upper():
                section = "Tendance Marche"
            elif "SECTEURS" in line.upper():
                section = "Performances Sectorielles"
            elif "TOP 5 HAUSSES" in line.upper():
                section = "Top Actions"
            elif "TOP 5 BAISSES" in line.upper():
                section = "Flop Actions"
            elif "MOYENNE PER" in line.upper() or "RENDEMENT" in line.upper():
                section = "Stats Globales"

            if ':' in line:
                parts = line.split(':', 1)
                field, value = parts[0].strip(), parts[1].strip()
                if section:
                    records.append((date, section, field, value))
                else:
                    records.append((date, "generic", field, value))

        return pd.DataFrame(records, columns=["Date", "Section", "Field", "Value"])

    def extract_table_data(self, date, raw_text):
        table = []
        pattern = re.compile(r'^(\w{3,5})\s+(\d+[.,]\d+)\s+([-+]?\d+[.,]\d+)%\s+([-+]?\d+[.,]\d+)%\s+(\d+[.,]\d+)?\s*(\d+[.,]\d+)?$')

        for line in raw_text.splitlines():
            match = pattern.match(line.strip())
            if match:
                ticker = match.group(1)
                price = float(match.group(2).replace(',', '.'))
                var_j = float(match.group(3).replace(',', '.'))
                var_y = float(match.group(4).replace(',', '.'))
                per = float(match.group(5).replace(',', '.')) if match.group(5) else None
                div = float(match.group(6).replace(',', '.')) if match.group(6) else None
                table.append((date, ticker, price, var_j, var_y, per, div))

        return pd.DataFrame(table, columns=["Date", "Ticker", "Price", "Variation_J", "Variation_Y", "PER", "Dividend"])

    def store_structured_info(self, df):
        conn = sqlite3.connect(self.db_path)
        df.to_sql("bulletin_data", conn, if_exists='append', index=False)
        conn.commit()
        conn.close()

    def store_table_data(self, df):
        conn = sqlite3.connect(self.db_path)
        df.to_sql("bulletin_table_data", conn, if_exists='append', index=False)
        conn.commit()
        conn.close()

    def process_new_bulletins(self):
        links = self.fetch_pdf_links()
        downloaded = self.download_bulletins(links)
        for date_str, fname in downloaded:
            path = os.path.join(self.storage_dir, fname)
            raw_text = self.parse_bulletin_text(path)
            structured_df = self.structure_info(date_str, raw_text)
            table_df = self.extract_table_data(date_str, raw_text)
            self.store_structured_info(structured_df)
            self.store_table_data(table_df)

    def load_bulletin_from_db(self, date):
        conn = sqlite3.connect(self.db_path)
        df = pd.read_sql("SELECT * FROM bulletin_data WHERE Date = ?", conn, params=(date,))
        conn.close()
        return df

    def load_table_data(self, date):
        conn = sqlite3.connect(self.db_path)
        df = pd.read_sql("SELECT * FROM bulletin_table_data WHERE Date = ?", conn, params=(date,))
        conn.close()
        return df
    def get_top_flop_actions(self, date):
        df = self.load_bulletin_from_db(date)
        top = df[df['Section'] == 'Top Actions'][['Field', 'Value']]
        flop = df[df['Section'] == 'Flop Actions'][['Field', 'Value']]
        return {'top': top, 'flop': flop}
    
    def get_sector_performance(self, date):
        df = self.load_bulletin_from_db(date)
        sector_df = df[df['Section'] == 'Performances Sectorielles']
        return sector_df[['Field', 'Value']].set_index('Field')['Value'].to_dict()

    def get_sector_performance(self, date):
        df = self.load_bulletin_from_db(date)
        sector_df = df[df['Section'] == 'Performances Sectorielles']
        return sector_df[['Field', 'Value']].set_index('Field')['Value'].to_dict()

    def get_price_summary_stats(self, date):
        df = self.load_table_data(date)
        return {
            'nb_stocks': len(df),
            'avg_price': round(df['Price'].mean(), 2),
            'avg_per': round(df['PER'].dropna().mean(), 2),
            'avg_dividend': round(df['Dividend'].dropna().mean(), 2),
            'avg_var_j': round(df['Variation_J'].mean(), 2),
            'avg_var_y': round(df['Variation_Y'].mean(), 2),
        }

    
    def extract_enhanced_action_table(self, date, raw_text):
        """
        Extrait une table combinée avec Ticker, Price, PER, Dividend, Volume, Bid/Ask info si disponible.
        """
        rows = []
        pattern = re.compile(
            r'^(\w{3,6})\s+(\d+[.,]\d+)\s+([-+]?\d+[.,]\d+)%\s+([-+]?\d+[.,]\d+)%'  # Ticker, Price, Var J/Y
            r'(?:\s+(\d+[.,]\d+))?'  # PER
            r'(?:\s+(\d+[.,]\d+))?'  # Dividend
            r'(?:\s+(\d+[.,]\d+))?'  # Bid
            r'(?:\s+(\d+[.,]\d+))?'  # Ask
            r'(?:\s+(\d+[.,]\d+))?'  # Volume
            r'$'
        )

        for line in raw_text.splitlines():
            match = pattern.match(line.strip())
            if match:
                vals = [match.group(i) for i in range(1, 10)]
                # Nettoyage des valeurs
                vals = [float(v.replace(',', '.')) if v else None for v in vals[1:]]
                rows.append((date, match.group(1)) + tuple(vals))

        return pd.DataFrame(rows, columns=[
            "Date", "Ticker", "Price", "Variation_J", "Variation_Y",
            "PER", "Dividend", "Bid", "Ask", "Volume"
        ])

    def store_enhanced_table(self, df):
        conn = sqlite3.connect(self.db_path)
        df.to_sql("bulletin_actions_full", conn, if_exists='append', index=False)
        conn.commit()
        conn.close()

        
    def load_enhanced_action_table(self, date):
        conn = sqlite3.connect(self.db_path)
        df = pd.read_sql("SELECT * FROM bulletin_actions_full WHERE Date = ?", conn, params=(date,))
        conn.close()
        return df
```

```python
import os
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import pandas as pd
import sqlite3
import re
from PyPDF2 import PdfReader

class BRVMBulletinManager:
    def __init__(self, storage_dir="../Bulletins", base_url="https://nsiafinance.com/donnees-de-marche/",
                 db_path="../DataBase/db_test.db"):
        self.storage_dir = storage_dir
        self.base_url = base_url
        self.db_path = db_path
        os.makedirs(self.storage_dir, exist_ok=True)
        # self._init_db()

    def _init_db(self):
        conn = sqlite3.connect(self.db_path)
        cursor = conn.cursor()
        cursor.execute('''
            CREATE TABLE IF NOT EXISTS bulletin_data (
                Date TEXT,
                Section TEXT,
                Field TEXT,
                Value TEXT,
                PRIMARY KEY (Date, Section, Field)
            )''')
        cursor.execute('''
            CREATE TABLE IF NOT EXISTS bulletin_actions_full (
                Date TEXT,
                Ticker TEXT,
                Price REAL,
                Variation_J REAL,
                Variation_Y REAL,
                PER REAL,
                Dividend REAL,
                Bid REAL,
                Ask REAL,
                Volume REAL,
                PRIMARY KEY (Date, Ticker)
            )''')
        conn.commit()
        conn.close()

    def fetch_pdf_links(self):
        response = requests.get(self.base_url)
        soup = BeautifulSoup(response.text, 'html.parser')
        links = [a['href'] for a in soup.find_all('a', href=True) if a['href'].lower().endswith('.pdf')]
        return [(self._extract_date_from_filename(link), os.path.basename(link), link) for link in links]

    def _extract_date_from_filename(self, filename):
        try:
            parts = filename.split('_')
            date_part = parts[-1].replace(".pdf", "")[:8]
            return datetime.strptime(date_part, "%Y%m%d").strftime("%Y-%m-%d")
        except:
            return None

    def download_bulletins(self, links):
        saved = []
        #links = self.base_url if links is None else links
        for date_str, fname, url in links:
            local_path = os.path.join(self.storage_dir, fname)
            if not os.path.exists(local_path):
                try:
                    res = requests.get(url)
                    with open(local_path, 'wb') as f:
                        f.write(res.content)
                    saved.append((date_str, fname))
                except Exception as e:
                    print(f"Download failed for {url}: {e}")
        return saved

    def parse_bulletin_text(self, filepath):
        reader = PdfReader(filepath)
        return "\n".join([p.extract_text() for p in reader.pages if p.extract_text()])

    def structure_info(self, date, text):
        section = None
        records = []
        for line in text.splitlines():
            line = line.strip()
            if not line:
                continue
            if "BRVM COMPOSITE" in line.upper(): section = "Tendance Marche"
            elif "SECTEURS" in line.upper(): section = "Performances Sectorielles"
            elif "TOP 5 HAUSSES" in line.upper(): section = "Top Actions"
            elif "TOP 5 BAISSES" in line.upper(): section = "Flop Actions"
            elif "MOYENNE PER" in line.upper() or "RENDEMENT" in line.upper(): section = "Stats Globales"
            if ':' in line:
                k, v = line.split(':', 1)
                records.append((date, section or 'generic', k.strip(), v.strip()))
        return pd.DataFrame(records, columns=["Date", "Section", "Field", "Value"])

    def extract_enhanced_action_table(self, date, text):
        rows = []
        pattern = re.compile(r'^(\w{3,6})\s+(\d+[.,]\d+)\s+([-+]?\d+[.,]\d+)%\s+([-+]?\d+[.,]\d+)%' +
                             r'(?:\s+(\d+[.,]\d+))?' +
                             r'(?:\s+(\d+[.,]\d+))?' +
                             r'(?:\s+(\d+[.,]\d+))?' +
                             r'(?:\s+(\d+[.,]\d+))?' +
                             r'(?:\s+(\d+[.,]\d+))?')
        for line in text.splitlines():
            match = pattern.match(line.strip())
            if match:
                ticker = match.group(1)
                nums = [float(x.replace(',', '.')) if x else None for x in match.groups()[1:]]
                rows.append((date, ticker) + tuple(nums))
        print(rows)
        return pd.DataFrame(rows, columns=[
            "Date", "Ticker", "Price", "Variation_J", "Variation_Y",
            "PER", "Dividend", "Bid", "Ask", "Volume"])

    def store_structured_info(self, df):
        conn = sqlite3.connect(self.db_path)
        df.to_sql("bulletin_data", conn, if_exists='append', index=False)
        conn.commit()
        conn.close()

    def store_enhanced_table(self, df):
        conn = sqlite3.connect(self.db_path)
        df.to_sql("bulletin_actions_full", conn, if_exists='append', index=False)
        conn.commit()
        conn.close()

    def process_new_bulletins(self):
        links = self.fetch_pdf_links()
        print(links)
        new_files = self.download_bulletins(links)
        for date_str, fname in new_files:
            if self.already_processed(date_str):
                print(f"BULLETIN {date_str} already processed, skipping.")
                continue
            path = os.path.join(self.storage_dir, fname)
            text = self.parse_bulletin_text(path)
            df_info = self.structure_info(date_str, text)
            df_table = self.extract_enhanced_action_table(date_str, text)
            self.store_structured_info(df_info)
            self.store_enhanced_table(df_table)

    def load_bulletin_from_db(self, date):
        conn = sqlite3.connect(self.db_path)
        df = pd.read_sql("SELECT * FROM bulletin_data WHERE Date = ?", conn, params=(date,))
        conn.close()
        return df

    def load_table_data(self, date):
        conn = sqlite3.connect(self.db_path)
        df = pd.read_sql("SELECT * FROM bulletin_actions_full WHERE Date = ?", conn, params=(date,))
        conn.close()
        return df
    
    def already_processed(self, date):
        conn = sqlite3.connect(self.db_path)
        cur = conn.cursor()
        cur.execute("SELECT COUNT(*) FROM bulletin_data WHERE Date = ?", (date,))
        result = cur.fetchone()[0]
        conn.close()
        return result > 0
```
